```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json("C:\Code\IT3190E_Group_32\Source Code\Data\preprocessed_movieDB.json")
```

```python
df = df[df['revenue']!=0]
df = df[df['budget']!=0]
```

```python
df
```

```python
print(df['release_date'].isnull().sum())
```

```python
df.dropna(subset=['release_date'], inplace=True)
```

```python
df.drop(columns = 'overview', inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

df['restriction'] = label_encoder.fit_transform(df['restriction'])
df.restriction.unique()
```

```python
df['genre'] = df['genre'].apply(lambda x: ', '.join(x))
df['directors'] = df['directors'].apply(lambda x: ', '.join(x) if x is not None else '')
df['actors'] = df['actors'].apply(lambda x: ', '.join(x) if x is not None else '')
df['release_date'] = df['release_date'].apply(lambda x: '-'.join(map(str, x)) if x is not None else '')
df
```

```python
df['release_date'] = pd.to_datetime(df['release_date'])
df['year'] = df['release_date'].dt.year.astype(int)
df['month'] = df['release_date'].dt.month.astype(int)
df.drop(columns = 'release_date', inplace = True)
```

```python
duration_mean = df['duration'].mean()
df['duration'] = df['duration'].replace(0, int(duration_mean))
```

```python
df
```

```python
actor_revenue_dict = {}

for index, row in df.iterrows():
    actors = row['actors'].split(', ')
    revenue = row['revenue']
    for actor in actors:
        if actor in actor_revenue_dict:
            actor_revenue_dict[actor] += revenue
        else:
            actor_revenue_dict[actor] = revenue

def calculate_star_power_actor(actors):
    actors_list = actors.split(", ")
    if not actors_list:
        return 0    
    revenues = [actor_revenue_dict.get(actor) for actor in actors_list]  
    return sum(revenues) / len(actors_list)  
```

```python
director_revenue_dict = {}

for index, row in df.iterrows():
    directors = row['directors'].split(', ')
    revenue = row['revenue']
    for director in directors:
        if director in director_revenue_dict:
            director_revenue_dict[director] += revenue
        else:
            director_revenue_dict[director] = revenue

def calculate_star_power_director(directors):
    directors_list = directors.split(", ")
    if not directors_list:
        return 0
    revenues = [director_revenue_dict.get(director) for director in directors_list]
    return sum(revenues) / len(directors_list)
```

```python
df['star_power_actors'] = df['actors'].apply(calculate_star_power_actor)
df.drop('actors', axis = 1, inplace = True)

df['star_power_directors'] = df['directors'].apply(calculate_star_power_director)
df.drop('directors', axis = 1, inplace = True)
```

```python
df['genre'] = df['genre'].str.lower()
genres = df['genre'].str.get_dummies(sep=', ')
genre_encoded = genres.add_prefix('genre_')

df = pd.concat([df, genre_encoded], axis=1)
df.drop(columns = 'genre', inplace = True)
```
